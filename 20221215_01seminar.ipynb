```python
#導関数定義
def numerical_diff(f,x):
    h=1e-4  
    return (f(x+h)-f(x)) / h
```

②関数xの差分について

>誤差が生じるということ



> **前提**

> 真の微分：xの位置での関数の傾き

> 近似による接線：（x＋h）とxの間の傾きに対応

> つまり、、真の微分と今回の実装値は、厳密には一致しない

> **改善ポイント**

> (x+h)と(x-h)での関数fの差分を計算する（中心差分）







```python
#導関数定義
def numerical_diff(f,x):
    h=1e-4  
    return (f(x+h)-f(x-h)) / (2*h)  #修正
```
